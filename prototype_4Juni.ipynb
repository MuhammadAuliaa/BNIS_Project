```python
import yfinance as yf
import plotly.graph_objects as go

def plot_stock_close_interactive(data, symbol):
    fig = go.Figure()

    # Menambahkan garis untuk harga penutupan
    fig.add_trace(go.Scatter(x=data.index, y=data['Close'], mode='lines', name='Close Price'))

    # Menambahkan garis untuk volume spike
    volume_spike = data[data['Volume'] > data['Volume'].quantile(0.95)]  # Filter volume yang lebih tinggi dari kuantil 95%
    fig.add_trace(go.Scatter(x=volume_spike.index, y=volume_spike['Close'], mode='markers', 
                             marker=dict(color='red', size=8), name='Volume Spike'))

    # Menambahkan judul dan label sumbu
    fig.update_layout(title=f'Stock Close Price ({symbol})',
                      xaxis_title='Date',
                      yaxis_title='Close Price',
                      template='plotly_dark')

    fig.show()

# Input simbol saham
stock_symbols = ['BBRI.JK', 'ASII.JK', 'TLKM.JK', 'BMRI.JK', 'MBMA.JK']

# Mengambil data saham dari Yahoo Finance untuk setiap simbol
for symbol in stock_symbols:
    stock_data = yf.download(symbol, start="2024-01-01", end="2024-06-03", interval='1d')
    plot_stock_close_interactive(stock_data, symbol)
```

```python
import yfinance as yf
import plotly.graph_objects as go

def plot_stock_interactive(data, symbol):
    fig = go.Figure()

    # Menambahkan garis untuk harga penutupan
    fig.add_trace(go.Scatter(x=data.index, y=data['Close'], mode='lines', name='Close Price'))

    # Menambahkan garis untuk volume
    fig.add_trace(go.Scatter(x=data.index, y=data['Volume'], mode='lines', name='Volume', yaxis='y2'))

    # Menandai volume spike dengan warna merah
    average_volume = data['Volume'].mean()
    volume_spike = data[data['Volume'] > 3 * average_volume]
    fig.add_trace(go.Scatter(x=volume_spike.index, y=volume_spike['Volume'], mode='markers', 
                             marker=dict(color='red', size=8), name='Volume Spike', yaxis='y2'))

    # Menambahkan judul dan label sumbu
    fig.update_layout(title=f'Stock Price and Volume ({symbol})',
                      xaxis_title='Date',
                      yaxis_title='Close Price',
                      yaxis2=dict(title='Volume', overlaying='y', side='right'),
                      template='plotly_dark')

    fig.show()

# Input simbol saham
stock_symbols = ['BBRI.JK', 'BBCA.JK', 'BBNI.JK', 'BMRI.JK', 'ASII.JK', 'TLKM.JK', 'SMGR.JK', 'INTP.JK', 'BBTN.JK', 'BRIS.JK', 'GGRM.JK', 'HMSP.JK', 'UNVR.JK']

# Mengambil data saham dari Yahoo Finance untuk setiap simbol
for symbol in stock_symbols:
    stock_data = yf.download(symbol, start="2024-01-01", end="2024-06-03", interval='1d')
    plot_stock_interactive(stock_data, symbol)
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Mengunduh data saham BBRI.JK dari Yahoo Finance
bbri = yf.download('BBRI.JK', start='2024-01-01', end='2024-06-03', interval='1d')

# Menampilkan 5 data pertama dari dataframe
print(bbri.head())

# Membuat line plot untuk atribut 'Close'
plt.figure(figsize=(10, 6))
plt.plot(bbri['Close'], label='Close Price', color='blue')
plt.title('BBRI.JK Close Price')
plt.xlabel('Date')
plt.ylabel('Price (IDR)')
plt.legend()
plt.grid(True)
plt.show()
```
