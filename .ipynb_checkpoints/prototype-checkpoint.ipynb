## Volume Spike

```python
def hitung_volume_spike(lot, freq_buy):
    volume_spike = lot * freq_buy * 10
    return volume_spike

# Contoh penggunaan
lot = 50000
freq_buy = 316

volume_spike = hitung_volume_spike(lot, freq_buy)
print("Volume spike:", volume_spike)
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Fungsi untuk menghitung volume spike
def calculate_volume_spike(data, threshold):
    # Menghitung rata-rata volume
    average_volume = data['Volume'].mean()
    # Menghitung volume spike
    spike_threshold = average_volume * threshold
    data['Volume Spike'] = data['Volume'] > spike_threshold
    return data

# Fungsi untuk memplot data saham beserta volume spike
def plot_stock_data(data):
    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data['Close'], label='Close Price')
    plt.bar(data.index, data['Volume'], color='gray', alpha=0.3, label='Volume')
    plt.scatter(data.index, data['Volume Spike']*data['Volume'].max(), color='red', marker='^', label='Volume Spike')
    plt.title('Stock Price and Volume with Volume Spike')
    plt.xlabel('Date')
    plt.ylabel('Price/Volume')
    plt.legend()
    plt.show()

# Input simbol saham
stock_symbol = input("Masukkan simbol saham (contoh: AAPL untuk Apple Inc.): ")

# Mengambil data saham dari Yahoo Finance
stock_data = yf.download(stock_symbol, start="2024-03-01", end="2024-06-01")

# Menghitung volume spike dengan threshold 2 (contoh)
stock_data = calculate_volume_spike(stock_data, 2)

# Memplot data saham beserta volume spike
plot_stock_data(stock_data)
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Fungsi untuk menghitung volume spike
def calculate_volume_spike(data, threshold_multiplier):
    # Menghitung rata-rata volume
    average_volume = data['Volume'].mean()
    # Menghitung threshold berdasarkan multiplier
    spike_threshold = average_volume * threshold_multiplier
    data['Volume Spike'] = data['Volume'] > spike_threshold
    return data

# Fungsi untuk memplot data saham beserta volume spike
def plot_stock_data(data):
    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data['Close'], label='Close Price')
    plt.bar(data.index, data['Volume'], color='gray', alpha=0.3, label='Volume')
    plt.scatter(data.index, data['Volume Spike']*data['Volume'].max(), color='red', marker='^', label='Volume Spike')
    plt.title('Stock Price and Volume with Volume Spike')
    plt.xlabel('Date')
    plt.ylabel('Price/Volume')
    plt.legend()
    plt.show()

# Input simbol saham
stock_symbol = input("Masukkan simbol saham (contoh: AAPL untuk Apple Inc.): ")

# Mengambil data saham dari Yahoo Finance
stock_data = yf.download(stock_symbol, start="2024-03-01", end="2024-06-01")

# Menghitung volume spike dengan threshold 10x rata-rata volume (contoh)
stock_data = calculate_volume_spike(stock_data, 3)

# Memplot data saham beserta volume spike
plot_stock_data(stock_data)
```

```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go

# Fungsi untuk menghitung volume spike
def calculate_volume_spike(data, threshold_multiplier):
    # Menghitung rata-rata volume
    average_volume = data['Volume'].mean()
    # Menghitung threshold berdasarkan multiplier
    spike_threshold = average_volume * threshold_multiplier
    data['Volume Spike'] = data['Volume'] > spike_threshold
    return data

# Fungsi untuk memplot data saham beserta volume spike (visualisasi interaktif)
def plot_stock_data_interactive(data):
    fig = go.Figure()

    # Menambahkan garis untuk harga penutupan
    fig.add_trace(go.Scatter(x=data.index, y=data['Close'], mode='lines', name='Close Price'))

    # Menambahkan batang untuk volume
    fig.add_trace(go.Bar(x=data.index, y=data['Volume'], name='Volume'))

    # Menambahkan markah untuk volume spike
    volume_spike_dates = data[data['Volume Spike']].index
    volume_spike_values = data[data['Volume Spike']]['Volume']
    fig.add_trace(go.Scatter(x=volume_spike_dates, y=volume_spike_values, mode='markers', 
                             marker=dict(color='red', symbol='triangle-up', size=10), name='Volume Spike'))

    # Menambahkan judul dan label sumbu
    fig.update_layout(title='Stock Price and Volume with Volume Spike',
                      xaxis_title='Date',
                      yaxis_title='Price/Volume')

    fig.show()

# Input simbol saham
stock_symbol = input("Masukkan simbol saham (contoh: AAPL untuk Apple Inc.): ")

# Mengambil data saham dari Yahoo Finance
stock_data = yf.download(stock_symbol, start="2024-03-01", end="2024-06-01")

# Menghitung volume spike dengan threshold 10x rata-rata volume (contoh)
stock_data = calculate_volume_spike(stock_data, 10)

# Memplot data saham beserta volume spike (visualisasi interaktif)
plot_stock_data_interactive(stock_data)
```

```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go

# Fungsi untuk menghitung volume spike
def calculate_volume_spike(data, threshold_multiplier):
    # Menghitung rata-rata volume
    average_volume = data['Volume'].mean()
    
    # Menghitung threshold berdasarkan multiplier
    spike_threshold = average_volume * threshold_multiplier
    data['Volume Spike'] = data['Volume'] > spike_threshold
    return data

# Fungsi untuk memplot data saham beserta volume spike (visualisasi interaktif)
def plot_stock_data_interactive(data, symbol):
    fig = go.Figure()

    # Menambahkan garis untuk harga penutupan
    fig.add_trace(go.Scatter(x=data.index, y=data['Close'], mode='lines', name='Close Price'))

    # Menambahkan batang untuk volume
    fig.add_trace(go.Bar(x=data.index, y=data['Volume'], name='Volume'))

    # Menambahkan markah untuk volume spike
    volume_spike_dates = data[data['Volume Spike']].index
    volume_spike_values = data[data['Volume Spike']]['Volume']
    fig.add_trace(go.Scatter(x=volume_spike_dates, y=volume_spike_values, mode='markers', 
                             marker=dict(color='red', symbol='triangle-up', size=10), name='Volume Spike'))

    # Menambahkan judul dan label sumbu
    fig.update_layout(title=f'Stock Price and Volume with Volume Spike ({symbol})',
                      xaxis_title='Date',
                      yaxis_title='Price/Volume')

    fig.show()

# Input simbol saham
stock_symbols = ['BBRI.JK', 'ASII.JK', 'TLKM.JK', 'BMRI.JK', 'MBMA.JK']

# Mengambil data saham dari Yahoo Finance untuk setiap simbol
for symbol in stock_symbols:
    stock_data = yf.download(symbol, start="2024-05-30", end="2024-06-03", interval='5m')
    stock_data = calculate_volume_spike(stock_data, 3)
    plot_stock_data_interactive(stock_data, symbol)
```
